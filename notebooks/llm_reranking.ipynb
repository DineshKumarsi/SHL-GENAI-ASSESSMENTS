```python
#Prepare Candidate Assessment (from Step 4)
candidates = df.iloc[indices[0]][
    ["name", "url", "test_type"]
].to_dict(orient="records")

candidates

#install OpenAI
!pip install openai

#Call LLM for re-ranking
from openai import OpenAI

client = OpenAI(api_key= "YOUR API KEY")

#Create LLM prompt

query = "Hiring a Java developer with good communication skills"

prompt = f"""
User query:
{query}

Candidate assessments:
{candidates}

Task:
- Select 5 to 10 assessments
- Ensure balance between:
  - Knowledge & Skills (K)
  - Personality & Behavior (P)
- Return only JSON in this format:
[
  {{
    "name": "",
    "url": ""
  }}
]
"""


#GET LLM RESPONSE

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

final_results = response.choices[0].message.content
final_results
```
